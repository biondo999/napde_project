```python
import pymesh
import numpy as np
import matplotlib.pyplot as plt

# Generate a triangular mesh in 2D
mesh = pymesh.generate_icosphere(1.0, [0.0, 0.0], refinement_order=2)

# Create a submesh by selecting a subset of faces
selected_faces = [1, 2, 5]  # Indices of the faces you want to select
submesh = pymesh.submesh(mesh, np.array(selected_faces))

# Visualize the original mesh in 2D
plt.figure(figsize=(8, 4))

plt.subplot(1, 2, 1)
plt.title("Original Mesh")
for face in mesh.faces:
    vertices = mesh.vertices[face]
    plt.plot(vertices[:, 0], vertices[:, 1], 'b-')
plt.axis('equal')

# Visualize the submesh in 2D
plt.subplot(1, 2, 2)
plt.title("Submesh")
for face in submesh.faces:
    vertices = submesh.vertices[face]
    plt.plot(vertices[:, 0], vertices[:, 1], 'r-')
plt.axis('equal')

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import tripy

# Generate random points for the mesh
np.random.seed(0)
points = np.random.rand(10, 2)

# Triangulate the points to create a 2D triangular mesh
triangles = tripy.earclip(points)

# Visualize the mesh
plt.figure(figsize=(6, 6))
for triangle in triangles:
    vertices = points[triangle]
    # Close the triangle for visualization
    closed_triangle = np.concatenate((vertices, [vertices[0]]))
    plt.plot(closed_triangle[:, 0], closed_triangle[:, 1], 'b-')

# Plot points
plt.plot(points[:, 0], points[:, 1], 'ro')

plt.title("2D Triangular Mesh")
plt.xlabel("X")
plt.ylabel("Y")
plt.axis('equal')
plt.show()

```

```python
import numpy as np
import matplotlib.pyplot as plt
import tripy

# Define the vertices of the square (counter-clockwise order)
square_vertices = np.array([[0, 0], [1, 0], [1, 1], [0, 1]])

# Create a grid of points inside the square
x_points = np.linspace(0, 1, 10)  # 10 points along the x-axis
y_points = np.linspace(0, 1, 10)  # 10 points along the y-axis
grid_points = np.array([[x, y] for x in x_points for y in y_points])

# Combine square vertices and grid points
all_points = np.vstack((square_vertices, grid_points))

# Triangulate the combined points to create a 2D triangular mesh
triangles = tripy.earclip(all_points)

# Visualize the mesh
plt.figure(figsize=(6, 6))
for triangle in triangles:
    vertices = np.array(triangle)
    # Close the triangle for visualization
    closed_triangle = np.concatenate((vertices, [vertices[0]]))
    plt.plot(closed_triangle[:, 0], closed_triangle[:, 1], 'b-')

# Plot square vertices
plt.plot(square_vertices[:, 0], square_vertices[:, 1], 'ro')

plt.title("2D Triangular Mesh on a Square Domain (Refined)")
plt.xlabel("X")
plt.ylabel("Y")
plt.axis('equal')
plt.show()
```

```python
import numpy as np
import pymesh
import matplotlib.pyplot as plt

# Define the vertices of the square (counter-clockwise order)
square_vertices = np.array([[0, 0], [1, 0], [1, 1], [0, 1]])

# Define the faces of the square
square_faces = np.array([[0, 1, 2], [0, 2, 3]])

# Create a PyMesh mesh object
mesh = pymesh.form_mesh(square_vertices, square_faces)

# Visualize the original mesh
plt.figure(figsize=(6, 6))
plt.triplot(mesh.vertices[:, 0], mesh.vertices[:, 1], mesh.faces, color='blue')
plt.plot(mesh.vertices[:, 0], mesh.vertices[:, 1], 'ro')
plt.title("Original Mesh")
plt.xlabel("X")
plt.ylabel("Y")
plt.axis('equal')
plt.show()

# Refine the mesh
refined_mesh = pymesh.subdivide(mesh, nsubdiv=3)

# Visualize the refined mesh
plt.figure(figsize=(6, 6))
plt.triplot(refined_mesh.vertices[:, 0], refined_mesh.vertices[:, 1], refined_mesh.faces, color='green')
plt.plot(refined_mesh.vertices[:, 0], refined_mesh.vertices[:, 1], 'ro')
plt.title("Refined Mesh")
plt.xlabel("X")
plt.ylabel("Y")
plt.axis('equal')
plt.show()
```

```python
import numpy as np
import pymesh
import matplotlib.pyplot as plt

# Define the vertices of the square (counter-clockwise order)
square_vertices = np.array([[0, 0], [1, 0], [1, 1], [0, 1]])

# Define the faces of the square
square_faces = np.array([[0, 1, 2], [0, 2, 3]])

# Create a PyMesh mesh object
mesh = pymesh.meshio.form_mesh(square_vertices, square_faces)

# Visualize the original mesh
plt.figure(figsize=(6, 6))
plt.triplot(mesh.vertices[:, 0], mesh.vertices[:, 1], mesh.faces, color='blue')
plt.plot(mesh.vertices[:, 0], mesh.vertices[:, 1], 'ro')
plt.title("Original Mesh")
plt.xlabel("X")
plt.ylabel("Y")
plt.axis('equal')
plt.show()

# Refine the mesh
refined_mesh = pymesh.meshutils.subdivide(mesh, order=1)

# Visualize the refined mesh
plt.figure(figsize=(6, 6))
plt.triplot(refined_mesh.vertices[:, 0], refined_mesh.vertices[:, 1], refined_mesh.faces, color='green')
plt.plot(refined_mesh.vertices[:, 0], refined_mesh.vertices[:, 1], 'ro')
plt.title("Refined Mesh")
plt.xlabel("X")
plt.ylabel("Y")
plt.axis('equal')
plt.show()
```
