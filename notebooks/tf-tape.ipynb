```python
@tf.function
def t(w):
    return w**2;
```

```python
@tf.function
def f(x0,x1,w):
    with tf.GradientTape() as g:
        g.watch(x1)
        y=x0*x1*t(w)
    grad=g.gradient(y,x1)
    return grad
```

```python
f(tf.constant(1.0),tf.constant(2.0),tf.constant(5.0))
```

```python
x0=tf.constant(1.0)
x1=tf.constant(2.0)
w=tf.Variable(5.0)

with tf.GradientTape() as tape:
    y=w*x0+f(x0,x1,w)
grad=tape.gradient(y,w)
print(grad)
```
