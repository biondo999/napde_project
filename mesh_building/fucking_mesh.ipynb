```python

class ref_triangle:
    def __init__(self,N):
        self.N=N
        self.h=1/(N-1)


    def print(self):
        for elem in range(0,(self.N+1)*(self.N)//2):
            self.print_single_elem(elem)
        

    def print_single_elem(self,elem):
            r,s=self.convert(elem)
            print("element number : ",elem," --> ",r,s,'-->>',self.convert_back(r,s))


    def convert(self,x):
         r=x
         n=self.N
         s=0
         while(r>=0 or n>0):
              if(r-n>=0):
                   r=r-n
                   s=s+1
                   n=n-1
              else:
                    break
              
         return r,s
    
    def convert_back(self,i,j):
        if j>0:
             return  sum(ii for ii in range(self.N-j+1,self.N+1))+i
        else:
             return i
        
              


    def get_coordinate(self,elem):
            q,r=divmod(elem, self.N) 
            x=np.float64(self.h*r)
            y=np.float64(self.h*q)
            return x,y,r,q
    
```

```python
T=ref_triangle(6)
```

```python
T.print()
```

```python

l=[]
for jj in range(0,T.N):
    for ii in range(0,T.N-jj):
        print(ii,jj,'triangle -->')

        temp=[]

        if(ii%2==0):
            print("even triangle")
            print(ii,jj+1)
            temp.append(T.convert_back(ii,jj+1))
            print(ii,jj)
            temp.append(T.convert_back(ii,jj))
            print(ii+1,jj)
            temp.append(T.convert_back(ii+1,jj))
        else:
            print("odd triangle")
            print(ii,jj+1)
            temp.append(T.convert_back(ii,jj+1))
            print(ii-1,jj+1)
            temp.append(T.convert_back(ii-1,jj+1))
            print(ii,jj)
            temp.append(T.convert_back(ii,jj))
        print(temp)

        l.append(temp)
      
```
