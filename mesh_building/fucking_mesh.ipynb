```python

class ref_triangle:
    def __init__(self,N):
        self.N=N
        self.h=1/(N-1)


    def print(self):
        for elem in range(0,(self.N+1)*(self.N)//2):
            self.print_single_elem(elem)
        

    def print_single_elem(self,elem):
            r,s=self.convert(elem)
            print("element number : ",elem," --> ",r,s,'-->>',self.convert_back(r,s))


    def convert(self,x):
         r=x
         n=self.N
         s=0
         while(r>=0 or n>0):
              if(r-n>=0):
                   r=r-n
                   s=s+1
                   n=n-1
              else:
                    break
              
         return r,s
    
    def convert_back(self,i,j):
        if j>0:
             return  sum(ii for ii in range(self.N-j+1,self.N+1))+i
        else:
             return i
        
              


    def get_coordinate(self,elem):
            q,r=divmod(elem, self.N) 
            x=np.float64(self.h*r)
            y=np.float64(self.h*q)
            return x,y,r,q
    
```

```python
def translate(nodes: list, vertices: list):
        """
        Translates given nodes in the reference
        case ([0, 0], [1, 0], [0, 1]) to an arbritrary triangle
        """

        # print(nodes)
        # print(vertices)

        output = np.zeros(shape=np.shape(nodes))

        output[:, 0] = (
            nodes[:, 0] * (vertices[1, 0] - vertices[0, 0])
            + nodes[:, 1] * (vertices[2, 0] - vertices[0, 0])
            + vertices[0, 0]
        )

        output[:, 1] = (
            nodes[:, 0] * (vertices[1, 1] - vertices[0, 1])
            + nodes[:, 1] * (vertices[2, 1] - vertices[0, 1])
            + vertices[0, 1]
        )

        det = (vertices[1, 0] - vertices[0, 0]) * (vertices[2, 1] - vertices[0, 1]) - (
            vertices[2, 0] - vertices[0, 0]
        ) * (vertices[1, 1] - vertices[0, 1])

        return output, det
```
